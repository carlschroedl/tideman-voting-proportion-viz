```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import random
from mpl_toolkits.mplot3d import Axes3D
from numpy import nan as nan
from IPython.display import Image
```

```python
from functools import reduce
import operator
def flatten(l):
    return reduce(operator.concat, l)

test_data = [[1,2,3], [4,5,6], [7], [8,9]]
assert flatten(test_data) == [1,2,3,4,5,6,7,8,9]
```

```python
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
# creating 3d plot using matplotlib 
# in python
  
# for creating a responsive plot
%matplotlib widget

fig = plt.figure()
ax = plt.axes(projection="3d")
x_len = 6
y_len = 5

# x entries are the beginning of the range for 
# each row in the table
# repeat each x entry y times
x_pos = flatten([ [value] * y_len for value in [3,5,7,10,15,20] ])

#     [3] * y_len,
#     [5] * y_len,
#     [7] * y_len,
#     [10]* y_len,
#     [15] * y_len,
#     [20] * y_len
# ])

# y entries are the begining of the range for
# each column in the table
# repeat each y entry x times
y_pos = [1, 2, 3, 5, 9] * x_len
    
num_bars = 30

z_pos = [0] * num_bars
# repeat each x entry y times
x_size = flatten([ [value] * y_len for value in [2,2,3,5,5,10]]) 
# repeat each y size entry x times
y_size = [1,1,2,4,4] * x_len

# z_size = [ random.random() for i in range(0,30)]

z_size = flatten([
    [.749, .783, .880, nan, nan],
    [.513, .604, .724, nan, nan],
    [.535, .476, .569, .799, nan],
    [.378, .383, .473, .481, .639],
    [nan, .312, .390, .480, nan],
    [nan, .234, nan, nan, .478]
])

# assign a pseudorandom color to each bar
seed = 1
my_colors = [ get_cmap(30)(i) for i in range(30)]
random.Random(seed).shuffle(my_colors)

ax.bar3d(x_pos, y_pos, z_pos, x_size, y_size, z_size, 
         color=my_colors
)

ax.set_title("Average Proportion of Candidates Ranked by Voters")
ax.set_xlabel('Number of Candidates')
ax.set_ylabel('Number Elected')
ax.set_zlabel('Average Proportion')
ax.set_zlim(0,1)
plt.show()
Image(filename='tideman-ranking-proportion.png') 
```
